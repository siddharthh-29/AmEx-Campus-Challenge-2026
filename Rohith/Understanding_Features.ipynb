# Understanding Features


## Cancellation Calls in last 12m. 

A binary flag (0/1) — whether the customer called to cancel their Premier card in the last 12 months

Domain interpretation
This is a genuinely distinct customer type from what f1/f3/f11 capture. Someone who calls to cancel is almost never a credit-risk case — their revolve balance and risk score are both dramatically lower than the base population, and they almost never overlap with collections. So this isn't "customer in financial trouble" — it looks like dissatisfaction-driven attrition: financially healthy customers who are unhappy with the value proposition (annual fee too high vs. benefits used, a bad service experience, a competing card offer) and are trying to leave.

## Average Revolver Balance in last 12m ##

What "revolve balance" means, conceptually
Credit card: You get a credit limit. If you don't pay in full, the unpaid amount "revolves" to next month and the issuer charges interest on it. Interest income is often the primary revenue driver for a classic revolving-credit issuer — so a bigger revolve balance = more interest revenue = good for the issuer (up to the point where default risk eats the gain).
Charge card: By design, you're expected to pay the full balance every month. No preset spending limit, and traditionally no interest-bearing revolve mechanic at all — Amex's classic charge card model. The deck confirms Premier is explicitly this: "No preset spending limit (charge card)."

But Amex Offers 

"Plan It" / Pay-Over-Time features — the deck itself lists "Plan It: installment flexibility" under Premier's digital experience. Modern Amex charge cards let customers convert big purchases into fixed installment plans, and Amex charges a monthly plan fee for this — meaning some revolve balance is intentional, fee-generating, and not a sign of distress at all.
Simple non-payment / delinquency — a customer just didn't pay their statement in full. On a true charge card this isn't a designed interest-revenue mechanic; it's a violation of the product's terms, which is why it's tightly linked to collections calls and risk flags rather than to profitability.

So: the dataset is still charge-card economics — it's not secretly a hybrid credit card product. The nonzero balances you're seeing are the byproduct of two very different customer behaviors that happen to look the same in a single number.
